### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
data = df[['DatabaseWantToWorkWith']].dropna()

#split database 
data["Database"] = data["DatabaseWantToWorkWith"].str.split(";")
df_exploded = data.explode("Database")

top5_database = df_exploded['Database'].value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_database, labels=top5_database.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
data = df[['DevType']].dropna()

#split database 
data["role"] = data["DevType"].str.split(";")
df_exploded = data.explode("role")

top5_role = df_exploded['role'].value_counts().nlargest(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_role, labels=top5_role.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 developer roles')
plt.show()
```
