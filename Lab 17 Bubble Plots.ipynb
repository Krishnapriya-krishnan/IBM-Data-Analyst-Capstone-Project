### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
df_data = df[['Age','SOPartFreq','JobSat']].dropna()
df_data.head()
bub_data = df_data.groupby(['Age','SOPartFreq'])['JobSat'].mean().reset_index()
bub_data.head()
plt.Figure(figsize=(12,6))
plt.scatter(
    bub_data["Age"],
    bub_data["SOPartFreq"],
    s=bub_data["JobSat"] * 5,   # scale bubble size
    alpha=0.6
)

plt.xlabel("Age Group")
plt.ylabel("Participation Frequency (SOPartFreq)")
plt.title("Age vs Participation Frequency\nBubble Size = Average Job Satisfaction")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
df_bubble = df[['ConvertedCompYearly','JobSat','Age']].dropna()
age_map = {
    "Under 18 years old": 16,
    "18-24 years old": 21,
    "25-34 years old": 30,
    "35-44 years old": 40,
    "45-54 years old": 50,
    "55-64 years old": 60,
    "65 years or older": 70
}

df_bubble["AgeNumeric"] = df_bubble["Age"].map(age_map)
#bub_data = df_data.groupby(['ConvertedCompYearly''Age','SOPartFreq'])['JobSat'].mean().reset_index()

plt.figure(figsize=(10,6))

plt.scatter(
    df_bubble["ConvertedCompYearly"],
    df_bubble["JobSat"],
    s=df_bubble["AgeNumeric"] * 3,   # scale bubble size
    alpha=0.6
)

plt.xlabel("Yearly Compensation")
plt.ylabel("Job Satisfaction")
plt.title("Compensation vs Job Satisfaction\nBubble Size Represents Age")
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here

df_bubble = df[['LanguageHaveWorkedWith','Age']].dropna()
# Split languages and explode into separate rows
df_bubble["Language"] = df_bubble["LanguageHaveWorkedWith"].str.split(";")
df_exploded = df_bubble.explode("Language")
df_grouped = (
    df_exploded
    .groupby(["Age", "Language"])
    .size()
    .reset_index(name="Frequency")
)
plt.figure(figsize=(20,20))

plt.scatter(
    df_grouped["Age"],
    df_grouped["Language"],
    s=df_grouped["Frequency"],
    alpha=0.6
)

plt.xlabel("Age Group")
plt.ylabel("Programming Language")
plt.title("Technology Preferences by Age Group\nBubble Size Represents Language Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python

##Write your code here

df_bubble = df[['DatabaseWantToWorkWith','JobSat']].dropna()
# Split languages and explode into separate rows
df_bubble["Database"] = df_bubble["DatabaseWantToWorkWith"].str.split(";")
df_exploded = df_bubble.explode("Database")
df_grouped = (
    df_exploded
    .groupby("Database")
    .agg(
        AvgJobSat=("JobSat", "mean"),
        Count=("JobSat", "count")
    )
    .reset_index()
)
plt.figure(figsize=(20,20))

plt.scatter(
    df_grouped["Database"],
    df_grouped["AvgJobSat"],
    s=df_grouped["Count"],
    alpha=0.6
)

plt.xlabel("Database")
plt.ylabel("job Satisfaction")
plt.title("Preferred Databases vs. Job Satisfaction\nBubble Size Represents Database Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
df_bubble = df[['ConvertedCompYearly','JobSat','DevType']].dropna()
# Split languages and explode into separate rows
df_bubble["developer_role"] = df_bubble["DevType"].str.split(";")
df_exploded = df_bubble.explode("developer_role")
df_grouped = (
    df_exploded
    .groupby("developer_role")
    .agg(
        Avgcomp=("ConvertedCompYearly", "mean"),
        Avgjobsat=("JobSat", "mean")
    )
    .reset_index()
)
df_grouped.head()
```

```python
plt.figure(figsize=(14,6))

plt.scatter(
    df_grouped["developer_role"],
    df_grouped["Avgcomp"],
    s=df_grouped["Avgjobsat"]*10,
    alpha=0.6
)

plt.xlabel("Developer Role")
plt.ylabel("Yearly Compensation")
plt.title("Compensation Across Developer Roles\nBubble Size Represents job satisfaction")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
df_bubble = df[['NEWCollabToolsHaveWorkedWith','Age']].dropna()
# Split NEWCollabToolsHaveWorkedWith and explode into separate rows
df_bubble.head()
df_bubble["CollabTool"] = df_bubble["NEWCollabToolsHaveWorkedWith"].str.split(";")
df_exploded = df_bubble.explode("CollabTool")

#Group data

df_grouped = (
    df_exploded
    .groupby(["Age", "CollabTool"])
    .size()
    .reset_index(name="Frequency")
)
plt.figure(figsize=(20,15))

plt.scatter(
    df_grouped["Age"],
    df_grouped["CollabTool"],
    s=df_grouped["Frequency"],   # scale bubble size
    alpha=0.6
)

plt.xlabel("Age Group")
plt.ylabel("Collaboration Tool")
plt.title("Collaboration Tools Used by Age Group\nBubble Size Represents Usage Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
df_bubble = df[['WebframeWantToWorkWith','JobSat']].dropna()
df_bubble.head()

#Split WebframeWantToWorkWith
df_bubble["WebFrame"] = df_bubble["WebframeWantToWorkWith"].str.split(";")
df_exploded = df_bubble.explode("WebFrame")

#Group data

df_grouped = (
    df_exploded
    .groupby([ "WebFrame"])
    .agg(
        AvgJobSat=("JobSat", "mean"),
        Count=("JobSat", "count")
    )
    .reset_index()
)
plt.figure(figsize=(20,15))

plt.scatter(
    df_grouped["WebFrame"],
    df_grouped["AvgJobSat"],
    s=df_grouped["Count"],   # scale bubble size
    alpha=0.6
)

plt.xlabel("Web Framework")
plt.ylabel("Average Job Satisfaction")
plt.title("Preferred Web Frameworks vs. Job Satisfaction\nBubble Size Represents number of respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.
