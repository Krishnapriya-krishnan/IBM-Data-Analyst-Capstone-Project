### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
data = df['ConvertedCompYearly'].dropna().reset_index()

plt.figure(figsize=(8,5))
plt.hist(data['ConvertedCompYearly'], bins=30)
plt.yscale('log')

plt.title('Distribution of Yearly Compensation')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Number of Respondents')
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
data = df['Age'].dropna().reset_index()
data.tail(20)

# Example age mapping
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64  years old': 59.5,
    '65 or older': 70
}

# Convert Age to numeric
data['AgeNumeric'] = data['Age'].map(age_mapping)

plt.figure(figsize=(8,5))
plt.hist(data['AgeNumeric'], bins=30)

plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
data = df[['Age','ConvertedCompYearly']].dropna().reset_index()

# Example age mapping
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64  years old': 59.5,
    '65 or older': 70
}

# Convert Age to numeric
data['AgeNumeric'] = data['Age'].map(age_mapping)
grouped_data = data.groupby('AgeNumeric')['ConvertedCompYearly'].mean().reset_index()
plt.figure(figsize=(8,5))
plt.scatter(
    grouped_data['AgeNumeric'],
    grouped_data['ConvertedCompYearly'],
    alpha=0.6
)

plt.title('Relationship Between Age and Yearly Compensation')
plt.xlabel('Age (Numeric)')
plt.ylabel('ConvertedCompYearly')
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
## Write your code here
data = df[['Age','ConvertedCompYearly','JobSatPoints_6']].dropna().reset_index()

# Example age mapping
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64  years old': 59.5,
    '65 or older': 70
}

# Convert Age to numeric
data['AgeNumeric'] = data['Age'].map(age_mapping)

#bub_data = data.groupby(['Age'])['ConvertedCompYearly','JobSatPoints_6'].median().reset_index()

plt.Figure(figsize=(12,6))
plt.scatter(
    data["ConvertedCompYearly"],
    data["JobSatPoints_6"],
    s=data["AgeNumeric"] * 5,   # scale bubble size
    alpha=0.6
)

plt.xlabel('Converted Compensation (Yearly)')
plt.ylabel('Job Satisfaction Score')
plt.title('Bubble Plot of Compensation vs Job Satisfaction (Bubble Size = Age)')

plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
data = df['MainBranch'].dropna().reset_index()
mainbranch_counts = data['MainBranch'].value_counts()
plt.figure(figsize=(10, 6))

plt.barh(mainbranch_counts.index, mainbranch_counts.values)

plt.xlabel('Number of Respondents')
plt.ylabel('Primary Role (MainBranch)')
plt.title('Distribution of Respondents by Primary Role')

plt.gca().invert_yaxis()  # Highest count at the top

plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
data = df['LanguageWantToWorkWith'].dropna().reset_index()
language_counts = data['LanguageWantToWorkWith'].value_counts().nlargest(5)
plt.figure(figsize=(10, 6))

plt.bar(language_counts.index, language_counts.values)

plt.xlabel('Programming Languages')
plt.ylabel('Top 5 Programming Languages Respondents Want to Work With')
plt.title('Distribution of Respondents by Top 5 Programming Languages')

plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
data = df[['JobSatPoints_6','JobSatPoints_7','Age']].dropna().reset_index()

median_job_sat = (
    data.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']]
    .median()
)


median_job_sat.plot(kind="bar",stacked=True,figsize=(10, 6))
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction Score')
plt.title('Stacked Bar Chart of Median Job Satisfaction by Age Group')
plt.legend()

plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
data = df['DatabaseHaveWorkedWith'].dropna().reset_index()
database_counts = data['DatabaseHaveWorkedWith'].value_counts().nlargest(15)
plt.figure(figsize=(10, 6))

plt.bar(database_counts.index, database_counts.values)

plt.xlabel('Number of Respondents')
plt.ylabel('Database Respondents Want to Work With')
plt.title('Distribution of Respondents by Database')
plt.xticks(rotation=90)

plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
data = df['ConvertedCompYearly'].dropna().reset_index()

median_comp = (
    df.groupby('Age')['ConvertedCompYearly']
    .median()
)

plt.figure(figsize=(10, 6))

plt.bar(
    median_comp.index,
    median_comp.values
)

plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (Yearly)')
plt.title('Median Yearly Compensation by Age Group')
plt.xticks(rotation=90)

plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
data = df['Country'].dropna().reset_index()

grouped_data = data.groupby('Country').size().reset_index(name='RespondentCount')
grouped_data.head()
grouped_data = grouped_data.sort_values(
    by='RespondentCount',
    ascending=False
)

plt.Figure(figsize=(12,6))

plt.bar(grouped_data['Country'], grouped_data['RespondentCount'])

plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.title('Distribution of Respondents by Country')
plt.xticks(rotation=90)
plt.show()
```
